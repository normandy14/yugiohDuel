```python
import pandas as pd
```

```python
df = pd.read_csv('YugiSheets.csv')
```

```python
df2 = df[['English name']]
```

```python
df2
```

```python
cardNames = []
for row in df2['English name']:
    cardNames.append(row.replace('"',''))
print (cardNames)
```

```python
df3 = pd.DataFrame(cardNames)
```

```python
df3.to_csv("YugiListClean.csv", header=False, sep=',',index=False)
```
